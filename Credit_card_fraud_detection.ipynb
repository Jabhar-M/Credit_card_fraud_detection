```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
# distribution of True transactions & Fraud transactions

df['Class'].value_counts()
```

```python
True_Trans = df[df.Class == 0]
Fraud_Trans = df[df.Class == 1]
```

```python
True_Trans.Amount.describe()
```

```python
Fraud_Trans.Amount.describe()
```

```python
#Finding Average amount of value for True and Fraud Transaction

df.groupby('Class').Amount.mean()
```

```python
# True Transaction is More Higher than Fradulant Trans SO we are going balancing the Dataset here.

True_Trans_sample = True_Trans.sample(n=492)
```

```python
New_df = pd.concat([True_Trans_sample, Fraud_Trans], axis=0 )
```

```python
#Here is the New Data frame which one is Perfectly Balanced Fraud and True Transaction

New_df
```

```python
# For Cross Checking

New_df['Class'].value_counts()
```

```python
# We are going to Seprate X and Y value for Predicition from Existing DF

X = New_df.drop(columns='Class', axis=1)
Y = New_df['Class']
```

```python
X
```

```python
Y
```

```python
#Using Train Split Data Techniques

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3)
```

```python
#Using K-fold Method to Find out Best Classification Algorithm

from sklearn.model_selection import cross_val_score

cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X, Y,cv=10)
```

```python
cross_val_score(SVC(gamma='auto'), X, Y,cv=10)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=20),X, Y,cv=10)
```

```python
#Finded the Better Algorithm wich one Predict Perfectly
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
```

```python
log.score(X_train, y_train)
```

```python
y_predicted =log.predict(X_test)
y_predicted
```
